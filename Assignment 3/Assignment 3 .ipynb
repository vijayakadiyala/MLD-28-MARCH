##### 1.1 Write a Python program to implement your own myreduce() function which works exactly like Python's built-in function reduce() 



```python
#Custom Reduce Function
def myreduce(func,inputvalues):      
    result=inputvalues[0]                  #intialize result with first item of your input list
    for i in inputvalues[1:]:              #run for loop for all items in list except first item 
        result=func(i,result)              #value of result will be updated with output of func() after every loop 
    return result

#sample function to check myreduce function execution
#Function to get sum of 2 numbers 
def sum(x,y):
    return x+y

#sample function to check myreduce function execution
#define function to get min of 2 numbers
def min(x,y):
    if x<y:
        return x
    else:
        return y
    
    
    
#sample function to check myreduce function execution
#define function to get max from 2 numbers 
def max(x,y):
    if x>y:
        return x
    else:
        return y

 
nums=[11, 3, 5, 6, 8, 3, 111, 5, 9, 987, 11, 222,456,333, 43, 2, 4, 6, 7,23456, 5543, 55, 22]    
    
from functools import reduce
print('Output using custom reduce function for sum of items in list:', myreduce(sum,nums))
print('Output using In-Built reduce function for sum of items in list:',reduce(lambda x,y:x+y,nums))
print('Output using custom reduce function for sum of items in list:', myreduce(max,nums))
print('Output using In-Built reduce function for sum of items in list:',reduce(lambda x,y:x if x>y else y,nums))
print('Output using custom reduce function for sum of items in list:', myreduce(min,nums))
print('Output using In-Built reduce function for sum of items in list:',reduce(lambda x,y:x if x<y else y,nums))
```

##### 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter() 


```python
# Custom Filter Function

def myfilter(func,inputvalues):
    result=[]
    for i in inputvalues:          # for every item in given list function is executed
        if func(i):                # if result of function is true next line willbe executed
            result.append(i)       # appends with item i everytime when if condition is met
    return result


# sample function to check myfilter function execution
# function to know if number is even or not
def is_even(i):
    if i%2==0:
        return True
    else:
        return False
    

#sample function to check myfilter function execution
# function to know if letter is vowel or not 
def is_vowel(i):
    vowels=['a','e','i','o','u','A','E','I','O','U']
    if i in vowels:
        return True
    else:
        False
 
alpha=['a','w','E','y','A']

 
nums=[1, 3, 5, 6, 8, 3, 1, 5, 9, 987, 11, 222, 333, 43, 2, 4, 6,23456,7, 5543, 55, 22]


print('Output using custom filter function to find all even numbers in list:', myfilter(is_even,nums))
print('Output using In-Built filter function find all even numbers in list:',list(filter(lambda x:x%2==0,nums)))
print('Output using custom filter function to find all vowels in list', myfilter(is_vowel,alpha))
print('Output using In-Built filter function to find all vowels in list:',list(filter(lambda x:x in 'aeiouAEIOU',alpha)))
```
