Write a function so that the columns of the output matrix are powers of the input vector. 
 
The order of the powers is determined by the increasing boolean argument. Specifically, when increasing is False, the i-th output column is the input vector raised element-wise to the power of N - i - 1. 
 
 
HINT: Such a matrix with a geometric progression in each row is named for AlexandreTheophile Vandermonde. 

```python
import numpy as np

def Vandermonde(in_matrix, n):
    m=len(in_matrix)
    out_matrix = np.array([x**(n-1-i) for x in in_matrix for i in range(n)]).reshape(m,n)
    return out_matrix

in_matrix=np.array([1,2,3,4,5])
n=5
Vandermonde(in_matrix,n)
```

```python
def moving_avg(array,window):
    for i in range(len(array)-window+1): # for loop over range of length of given array - windowsize
       # print(sum(array[i:window+i]))
        print('y'+str(i+1)+' = '+str(sum(array[i:window+i])/window))  #for every loop it execute sum of 3 items which divided y window size
    
a=np.array([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])

moving_avg(a,3)
        
```

```python
def moving_avg1(array,n):
    a=np.cumsum(array)
    a[n:]=a[n:]-a[:-n]
    return (a[n-1:]/n)
    
moving_avg1(a,3)    
```
